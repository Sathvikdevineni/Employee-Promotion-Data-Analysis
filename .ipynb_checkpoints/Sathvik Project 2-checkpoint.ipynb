# Decision Tree algorithm

vector feature and target variable

```python
X1 = df2.drop(['is_promoted'], axis=1)

y1 = df2['is_promoted']
```

```python
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size = 0.2, random_state = 42)
```

```python
# check the shape of X_train and X_test

X_train.shape, X_test.shape
```

```python
# Feature Engineering 
# check data types in X_train

X_train.dtypes
```

```python
#Encode categorical variables
X_train.head()
```

```python
import category_encoders as ce
```

```python
encoder = ce.OrdinalEncoder(cols=['department', 'region', 'education', 'gender','recruitment_channel', 'no_of_trainings', 'age','previous_year_rating','length_of_service','awards_won','avg_training_score'])


X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

X_train.head()

```python
X_test.head()
```

Decision Tree Classifier with criterion gini index

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
crit_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)


# fit the model
crit_gini.fit(X_train, y_train)
```

```python
#Predict the Test set results with criterion gini index
y_pred_gini = crit_gini.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
```

```python
#Now, I will compare the train-set and test-set accuracy to check for overfitting
y_pred_train_gini = crit_gini.predict(X_train)

y_pred_train_gini
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(crit_gini.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(crit_gini.score(X_test, y_test)))
```

These two values are quite comparable. So, there is no sign of overfitting.

```python
#Visualize decision-trees
plt.figure(figsize=(16,10))

from sklearn import tree

tree.plot_tree(crit_gini.fit(X_train, y_train)) 
```

 Decision Tree Classifier with criterion entropy

```python
crit_en = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)


# fit the model
crit_en.fit(X_train, y_train) 
```

Predict the Test set results with criterion entropy

```python
y_pred_en = crit_en.predict(X_test)
```

```python
#Check accuracy score with criterion entropy

from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion entropy: {0:0.4f}'. format(accuracy_score(y_test, y_pred_en)))
```

```python
# I will compare the train-set and test-set accuracy to check for overfitting
y_pred_train_en = crit_en.predict(X_train)

y_pred_train_en
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_en)))
```

```python
#Check for overfitting and underfitting
# print the scores on training and test set

print('Training set score: {:.4f}'.format(crit_en.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(crit_en.score(X_test, y_test)))
```

```python
plt.figure(figsize=(16,10))

from sklearn import tree

tree.plot_tree(crit_en.fit(X_train, y_train)) 
```

Confusion matrix

```python
# Print the Confusion Matrix and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_en)

print('Confusion matrix\n\n', cm)
```

```python
#classification report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_en))
```
